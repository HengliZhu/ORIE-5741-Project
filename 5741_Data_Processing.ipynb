```python
import pandas as pd
import yfinance as yf
import numpy as np



#we do not need this function
#here is a function to get info of a single stock for a time period
#daily adj close and volume as column. Also ticker, sector and FIGI as unique identifier
def get_individual_data(ticker, start_date, end_date):
    
    stock_data = yf.download(ticker, start=start_date, end=end_date)
    
    # Keep only 'Adj Close' and 'Volume', and ensure the 'Date' is a column
    stock_data.reset_index(inplace=True)
    stock_data = stock_data[['Date', 'Adj Close', 'Volume']]
    
    global stock_info
    
    additional_info = stock_info[stock_info['Ticker'] == ticker]
    
    if not additional_info.empty:
        additional_info = additional_info.iloc[0]
        stock_data.loc[:, 'FIGI'] = additional_info['FIGI']
        stock_data.loc[:, 'Sector'] = additional_info['Sector']
    else:
        stock_data.loc[:, 'FIGI'] = 'N/A'
        stock_data.loc[:, 'Sector'] = 'N/A'
    
    return stock_data

# Define the CSV data once, outside the function
csv_path ="/Users/fuxiyao/Desktop/康奈尔Cornell/2024 Spring/messy Data/5741 project/russell3000_modified.csv" 
stock_info = pd.read_csv(csv_path) #file: russell3000_universe.csv

# Example usage:
df = get_individual_data('JPM', '2020-01-01', '2022-12-31')
df
```

```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('/Users/fuxiyao/Desktop/康奈尔Cornell/2024 Spring/messy Data/5741 project/russell3000_price.csv')

# Pivot the DataFrame to have PERMNOs as rows and dates as columns
#rows be each stock indntify by their unique permno, 
#for columns, daily price (PRC) for daily in the dataset.
df_pivot = df.pivot(index='PERMNO', columns='date', values='PRC')

# Fill missing values with NaN
df_pivot.fillna(float('nan'), inplace=True)

```

```python
df_pivot
```

```python
output_filename = 'russell3000_price_pivoted.csv'
df_pivot.to_csv(output_filename, index=False)

print(f"DataFrame saved as {output_filename}")
```

```python
threshold = int(0.9999 * len(df_pivot.columns))

# Remove rows (stocks) where more than 0.5% of the values are NaN
df_pivot_filtered = df_pivot.dropna(thresh=threshold)

df_pivot_filtered
```

```python
#Use Time Weighted interpolation to impute missing values 
#method that preserves the overall patterns and relationships in the data,with Non-uniform time 
#while minimizing the impact of missing values on the clustering process.


# Convert the date column to datetime format
df_pivot_filtered.columns = pd.to_datetime(df_pivot_filtered.columns)

# Transpose the DataFrame
df_pivot_filtered_transposed = df_pivot_filtered.transpose()

# Perform time-weighted interpolation
df_pivot_filtered_imputed = df_pivot_filtered_transposed.interpolate(method='time')

# Transpose the DataFrame back to its original orientation
df_pivot_filtered_imputed
```

```python
output_filename = 'russell3000_price_clean.csv'
df_pivot_filtered_imputed.to_csv(output_filename, index=True)

print(f"DataFrame saved as {output_filename}")
```

```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('/Users/fuxiyao/Desktop/康奈尔Cornell/2024 Spring/messy Data/5741 project/russell3000_price.csv')

# Read the Excel file containing stock information
stock_info = pd.read_csv('/Users/fuxiyao/Desktop/康奈尔Cornell/2024 Spring/messy Data/5741 project/russell3000_universe.csv')  

sectors_to_keep = ['Financials', 'Energy']
filtered_tickers = stock_info[stock_info['Sector'].isin(sectors_to_keep)]['Ticker'].unique()
df_filtered = df[df['TICKER'].isin(filtered_tickers)]

# Pivot the DataFrame to have PERMNOs as rows and dates as columns
df_pivot = df_filtered.pivot(index='PERMNO', columns='date', values='PRC')

# Fill missing values with NaN
df_pivot.fillna(float('nan'), inplace=True)

# Remove rows (stocks) where more than 0.5% of the values are NaN
threshold = int(0.995 * len(df_pivot.columns))
df_pivot_filtered = df_pivot.dropna(thresh=threshold)

# Convert the date column to datetime format
df_pivot_filtered.columns = pd.to_datetime(df_pivot_filtered.columns)

# Transpose the DataFrame
df_pivot_filtered_transposed = df_pivot_filtered.transpose()
```

```python
df_pivot_filtered_imputed = df_pivot_filtered_transposed.interpolate(method='time')

# Transpose the DataFrame back to its original orientation
df_pivot_filtered_imputed = df_pivot_filtered_imputed

# Save the filtered and imputed DataFrame to a new CSV file
output_filename = 'russell3000_price_financials_energy.csv'
df_pivot_filtered_imputed.to_csv(output_filename, index=True)
print(f"DataFrame saved as {output_filename}")
```
